```python
import pandas as pd
```

```python
# Load csv
filename = "/home/admin-tarassow/git/forecasting_principles/data/csv/vic_elec.csv"
data = pd.read_csv(filename)
print(data.head())
```

```python
# Column Time is a timestamp with value "2011-12-31T13:00:00Z". The dataset includes an observation for every 30 minutes. Please aggregate data to daily frequency. However, the following aggregation is wanted:
# 1) Demand = sum(Demand) / 1e3
# 2) Temperature = max(Temperature),
# 3) Holiday = any(Holiday)

data['Time'] = pd.to_datetime(data['Time'])
data.set_index('Time', inplace=True)
daily_data = data.resample('D').agg({
    'Demand': lambda x: x.sum() / 1e3,
    'Temperature': 'max',
    'Holiday': 'any'
})
```

```python
# Store column time as date only
daily_data.index = daily_data.index.date
daily_data.reset_index(inplace=True)
daily_data.rename(columns={'index': 'Date'}, inplace=True)
print(daily_data.head())

# Cast Holiday to integer
daily_data['Holiday'] = daily_data['Holiday'].astype(int)

# Save to new csv
filename = "/home/admin-tarassow/git/forecasting_principles/data/csv/vic_elec_daily.csv"
daily_data.to_csv(filename, index=False)
print(f"Daily aggregated data saved to {filename}")
```
